```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df = sns.load_dataset("mpg")
df
```

```python
df["origin"].unique()
```

```python
from sklearn.preprocessing import OneHotEncoder


ob1= OneHotEncoder()

encoded_data = ob1.fit_transform(df[["origin"]])
df1= pd.DataFrame(encoded_data.toarray(), columns=ob1.get_feature_names_out())

final_encoded= pd.concat([df,df1],axis=1)
print(final_encoded)
```

```python
final_encoded.drop("origin",axis=1,inplace=True)
```

```python
#DATA ENCODING BY LEBEL ENCODER

from sklearn.preprocessing import LabelEncoder

lebel= LabelEncoder()
df["encoded_origin"]=lebel.fit_transform(df[["origin"]])
print(df)

```

```python
from sklearn.model_selection import train_test_split

x= df.iloc[:,:-1]
y= df["encoded_origin"]
x_train,x_test,y_train, y_test= train_test_split(x,y, test_size=0.2,random_state=1)
print(x_train,x_test,y_train, y_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso

las= Lasso()
param_grid={"alpha":[0.1,0.001,0.5, 0.3,0.22]}
grid_search= GridSearchCV(estimator=las,param_grid=param_grid,scoring="r2", cv=5,verbose=2)

```

```python
grid_search.fit(x_train,y_train)
```
